```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
num_ats = pd.Series(np.hstack((np.tile(np.arange(2,9), 3), np.tile(np.arange(9,11), 3))))
```

```python
usamples = ["WT", "OB/OB", "WT", "HFD"]
lut = {usamples[i]: plt.get_cmap("tab10")(i) for i in range(4)}
```

```python
wt_loss = pd.concat((pd.read_csv('results/WT_num_archetypes_loss.txt', sep=' ', header=None), num_ats), axis=1)
wt_loss.columns = ['run', 'condition', 'MSE', 'Num archetypes']
hfd_loss = pd.concat((pd.read_csv('results/HFD_num_archetypes_loss.txt', sep=' ', header=None), num_ats), axis=1)
hfd_loss.columns = ['run', 'condition', 'MSE', 'Num archetypes']
obob_loss = pd.concat((pd.read_csv('results/OB_OB_num_archetypes_loss.txt', sep=' ', header=None), num_ats), axis=1)
obob_loss.columns = ['run', 'condition', 'MSE', 'Num archetypes']
```

```python
fig = plt.figure()
ax = sns.lineplot(data=wt_loss, x='Num archetypes', y='MSE', label='WT', color=lut['WT'])
fig.savefig('figures/WT_num_archetypes_loss.png', dpi=300)
```

```python
fig = plt.figure()
ax = sns.lineplot(data=hfd_loss, x='Num archetypes', y='MSE', label='HFD', color=lut['HFD'])
fig.savefig('figures/HFD_num_archetypes_loss.png', dpi=300)
```

```python
fig = plt.figure()
ax = sns.lineplot(data=obob_loss, x='Num archetypes', y='MSE', label='OB/OB', color=lut['OB/OB'])
fig.savefig('figures/OB_OB_num_archetypes_loss.png', dpi=300)
```
